```python
import numpy as np
import matplotlib.pyplot as plt

#  
a = 1.5  #     
theta = np.linspace(0, 2*np.pi, 1000)  # 0 2 1000  

#       
r1 = a * (1 - np.sin(theta))
r2 = a * (1 + np.sin(theta))

#       
fig, ax = plt.subplots(1, 1, subplot_kw={'projection': 'polar'})
ax.plot(theta, r1, color='red')  #    
ax.plot(theta, r2, color='red')  #    

#  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#  
t = np.linspace(-2*np.pi, 2*np.pi, 1000)
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

#  
plt.plot(x, y, color='gray')
plt.gca().set_aspect('equal', adjustable='box')  #     
plt.axis('off')  #  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import splprep, splev

#      
x = [0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0, 1.05, 1.1, 1.15, 1.2]
y = [0.5, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1.0, 1.05, 1.1, 1.15, 1.2, 1.25]

#        
tck, u = splprep([x, y], s=0)
new_points = splev(np.linspace(0, 1, 100), tck)

#  
plt.plot(new_points[0], new_points[1], color='gray')
plt.gca().set_aspect('equal', adjustable='box')  #     
plt.axis('off')  #  
plt.show()
```
